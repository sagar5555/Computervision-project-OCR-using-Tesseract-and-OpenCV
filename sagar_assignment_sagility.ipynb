```python
dates=[]
bank=[]
amount=[]
```

```python
text1=pytesseract.image_to_string(img1)
```

```python
l1=[i.split(' ') for i in text1.split('\n')]
l1
```

```python
dates.append(l1[1][6])
bank.append(l1[9][8]+' '+l1[9][9])
amount.append(l1[7][3])
```

```python
bank
```

```python
text2=pytesseract.image_to_string(img2)
```

```python
l2=[i.split(' ') for i in text2.split('\n')]
l2
```

```python
dates.append(l2[16][0])
amount.append(l2[21][1])
bank.append(l2[29][0].split(']')[0]+' '+l2[29][1])
```

```python
text3=pytesseract.image_to_string(img3)
```

```python
l3=[i.split(' ') for i in text3.split('\n')]
l3
```

```python
dates.append(l3[4][2])
amount.append(l3[19][1])
bank.append(l3[21][0]+' '+l3[21][1])
```

```python
text4=pytesseract.image_to_string(img4)
```

```python
l4=[i.split(' ') for i in text4.split('\n')]
l4
```

```python
dates.append(l4[2][6])
amount.append(l4[7][1])
bank.append(l4[9][2]+' '+l4[9][3])
```

```python
df=pd.DataFrame([dates,amount,bank]).transpose()
```

```python
df.columns=['dates','amount','bank_name']
```
